```python
import pandas as pd
print(pd.__version__)
```

```python
df = pd.read_csv("../data/raw/retail_sales_dataset.csv")
df.head()
df.info()
```

```python
# Standardize column names
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

df.head()
```

```python
df["date"] = pd.to_datetime(df["date"])

df.dtypes
```

```python
df["calculated_total"] = df["quantity"] * df["price_per_unit"]

(df["calculated_total"] != df["total_amount"]).sum()
```

```python
df_clean = df.drop(columns=["calculated_total"])

df_clean.to_csv(
    "../data/processed/cleaned_retail_sales.csv",
    index=False
)
```
